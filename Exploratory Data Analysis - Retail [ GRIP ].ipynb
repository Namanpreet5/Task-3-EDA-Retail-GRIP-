```python
#Visualizing Data :
#Importing seaborn and matplotlib libraries for the same.

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#Shipment Count across different Ship modes :
```

```python
plt.style.use('bmh')
sns.countplot(x='Ship Mode',data=ss)
plt.ylabel('No. of Shipments')
```

```python
#Standard Class was the most preferred Shipment Mode and Same Day was the least preferred option. 
```

```python
#Analyzing Segment :
```

```python
seg=ss['Segment'].value_counts()
colors=['blue','green','red']
explode=[0.1,0.0,0.0]
plt.pie(seg.values, labels=seg.index, autopct="%1.1f%%",colors=colors,explode=explode,startangle=60)
plt.title("Distribution of segment",fontsize=17)
plt.show()
```

```python
#Consumer segment has highest share followed by Corporate and Home Office.
```

```python
# Dropping columns that are of no use : 
```

```python
ss.drop(['Postal Code'],axis=1,inplace=True)
ss.drop(['Country'],axis=1,inplace=True)
```

```python
ss.columns
```

```python
#Analyzing Cities:

ss.City.value_counts().nlargest(20).plot(kind= 'bar',color='Aqua',width=0.8)
plt.title("Distribution of top 20 cities",fontsize=15)
plt.xlabel("Cities")
plt.ylabel("Count")
plt.show()
```

```python
#Analyzing Cities:

ss.City.value_counts().nsmallest(20).plot(kind= 'bar',color='purple',width=0.8)
plt.title("Distribution of bottom 20 cities",fontsize=15)
plt.xlabel("Cities")
plt.ylabel("Count")
plt.show()
```

```python
# Analyzing regions :

sns.histplot(ss.Region,color='blue',edgecolor='Orange')
plt.title("Distribution of region",fontsize=20)
plt.show()
```

```python
# West region have max customer
# We must focus on South region to improve our profit

# Analyzing categories :

sns.histplot(ss.Category,color='yellow',edgecolor='g')
plt.title("Distribution of Categories",fontsize=20)
plt.show()
```

```python
# Office Supplies have heighest count
# We must focus on Furniture as well as Technology to improve our sales
```

```python
#Plotting Pairwise relationship b/w columns
```

```python
sns.pairplot(data=ss, hue='Category')
```

```python
#HeatMap :
```

```python
ss_heat=ss.corr()
sns.heatmap(ss_heat, annot=True, cmap='Blues')
```

```python
#Analyzing Sales across different Categories :

plt.bar(ss['Category'],ss['Sales'],color = 'maroon' )
plt.xlabel('Cateories')
plt.ylabel('Sales')
plt.show()
```

```python
# Analyzing Count of Segment by Category and Segment :

sns.countplot(x='Category',hue='Segment',data=ss )
plt.ylabel('Count of Segment')
plt.title('Count of Segment by Category and Segment')
```

```python
#JointPlot depicting Sales across Regions :

sns.jointplot(x='Region', y='Sales', data=ss)
```

```python
#CountPlot depicting the No. of Sub-Categories

plt.figure(figsize=(16,9))

sns.countplot(ss['Sub-Category'], palette='dark')

plt.title('Count of Sub-Category', fontsize=10)
```

```python
#LinePlot depicting relation b/w Discount and Profit
sns.lineplot(y="Profit", x="Discount",data=ss,color='blue')
```

```python
#State-wise Distribution of Consumers

state_count=ss["State"].value_counts()
state_count
state_count.plot(kind="bar", figsize=(15,8))
plt.xlabel("States")
plt.ylabel("No. of Consumers")
plt.title("Distribution of Consumers in each State")
```

```python
#Analyzing Sales and Profit by different Sub-Categories
ss.groupby('Sub-Category')['Sales','Profit','Discount'].agg(sum).plot(kind = 'barh',color=['pink','brown','orange'])
plt.title('Total Sales and Profits by different Sub-Categories')
plt.ylabel('Total Sales and Profits')
plt.show()
```

```python
#Phones had the highest sales among all Sub-Categories followed by Chairs, Storage, Tables, etc. 
#Copiers made the highest profit. Bookcases made the least profit but made good Sales. The cost of Bookcases can be increased to increase the profit.
```

```python
#Summing up the Sales across various Categories:

category_sales_sum=ss.groupby("Category").Sales.sum()
category_sales_sum

label=['Furniture','Office Supplies','Technology']
plt.style.use('fivethirtyeight')
category_sales_sum.plot(kind="pie", figsize=(7,7),labels=label,autopct='%1.1f%%',fontsize=13, shadow=True)
```

```python
#Profit & Sales across various Categories :

ss.groupby('Category')['Sales','Profit'].agg(sum).plot(kind='bar')
plt.ylabel('Profit & Sales')
plt.show()
```

```python
#Segment-wise Profit & Sales:

seg_salesNprofit=ss.groupby("Segment")[["Sales", "Profit"]].sum()
seg_salesNprofit.plot(kind="bar", figsize=(8,5))
plt.ylabel("Sales & Profit")
plt.title("Segment: Profit & Sales", fontsize=15)
```

```python
#Summing up the Profit across different Regions:

region_profit_sum=ss.groupby("Region").Profit.sum()
region_profit_sum
```

```python
plt.style.use('bmh')
standout=[0,0,0,0.1]
plt2=region_profit_sum.plot(kind="pie",explode=standout, figsize=(7,7),autopct='%1.1f%%',fontsize=12, shadow=True)
```

```python
#Analyzing the sales across different Regions:
    
ss.groupby('Region')['Sales'].agg(sum).plot(kind = 'bar')
plt.title('Total Sales by different Regions')
plt.ylabel('Total Sales')
plt.xlabel('Regions')
plt.show()
```
